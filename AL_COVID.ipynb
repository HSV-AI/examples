# Alabama COVID Projections

This notebook is an attempt at evaluating various methods of projecting time series data. Currently the best available data for daily reported cases & deaths is [https://www.alreporter.com/mapping-coronavirus-in-alabama/](https://www.alreporter.com/mapping-coronavirus-in-alabama/).

The initial thought is to apply the following tutorial in Time Series Projection to the small AL COVID data set. [An End-to-End Project on Time Series Analysis and Forecasting with Python](https://towardsdatascience.com/an-end-to-end-project-on-time-series-analysis-and-forecasting-with-python-4835e6bf050b)

```
# This is the total cases reported in AL
cases = [6,12,22,29,39,51,78,106,131,157,196,242,440,531,639,720,830,947,999, \
         1106,1270,1535,1633,1841,2006,2197,2499,2838,3008,3262,3583]
```

```
import matplotlib.pyplot as plt
import pandas as pd

ts = pd.Series(cases, index=pd.date_range('3/13/2020', periods=len(cases)))
ax = ts.plot(figsize=(11,9))
ax.set_ylabel('Total Cases')
ax.set_xlabel('Date')
plt.show()
```
